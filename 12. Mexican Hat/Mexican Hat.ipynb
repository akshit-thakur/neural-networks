## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

## Utility function

```python
def activation_function(x,x_max=2):
    if x < 0:
        return 0
    elif x <= x_max:
        return x
    else:
        return x_max
```

## Mexican Hat

```python
s = [0.27, 0.35, 0.44, 0.58, 0.66, 0.77, 0.4, 0.32, 0.2, 0.15, 0.08]
#step 0
R1 = 1 #well-wisher neighbourhood
R2 = 3 #not well-wisher neighbourhood

n = len(s) #number of neurons

t_max = 10

C1 = 0.6 #weights of all well-wishers
C2 = -0.4 #weights of all not well-wishers

#step 1 
x = s 
t = 0 #time iteration

plt.figure(figsize=(12,12))

#step 2
while t < t_max:
    #step 3
    x_old = x #step 5
    x = []
    for i in range(n):
        summation = 0
        #finding well wishers contribution
        start = i-R1 
        end = i+R1
        if start < 0:
            start = 0
        if end > n:
            end = n
        for k in range(start,end):
            summation += x_old[k]
        summation_c1 = C1*summation
        
        #finding not well wishers contribution on the left side of neuron
        summation = 0
        start = i-R2
        end = i-R1-1
        if start < 0:
            start = 0
        if end < 0: #both negatives so no > condition check
            end = 0
        for k in range(start,end):
            summation += x_old[k]
        summation_c2_left = C2*summation
        
        #finding not well wishers contribution on the right side of neuron
        summation = 0
        start = i-R1+1
        end = i+R2
        if start < 0:
            start = 0
        if end > n:
            end = n
        summation_c2_right = C2*summation
        
        intermediate_output = summation_c1+summation_c2_left+summation_c2_right 
        x.append(activation_function(intermediate_output)) #step 5
    
    plt.plot(x,color = ListedColormap(('red', 'green','blue','orange','yellow','pink','brown','cyan','purple','crimson'))(t%10))
    
    t+=1 #step 6

#step 7    
plt.show()
```
