## Importing Libraries

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

## Reading dataset

```python
df = pd.read_csv('xor.csv')

x = df.iloc[:,[0,1]]
y = df.iloc[:,-1]
```

```python
plt.figure(figsize = (10, 6))
plt.title('XOR')
for i in range(len(df)):
    plt.scatter(x['x1'][i],x['x2'][i],color = ListedColormap(('red', 'green'))(y[i]))
plt.show()
```

```python
input_matrix = x.to_numpy().tolist()
```

```python
input_matrix
```

```python
def gaussian_rbf(input_row,receptors):
    x1,x2 = input_row
    t1,t2 = receptors
    r1 = (x1-t1[0])**2+(x2-t1[1])**2
    r2 = (x1-t2[0])**2+(x2-t2[1])**2
    return [round(math.exp(-r1),1),round(math.exp(-r2),1)]
```

```python
receptors = [[0,0],[1,1]]
output_matrix = []

for row in input_matrix:
    output_row = gaussian_rbf(row,receptors)
    output_matrix.append(output_row)
```

```python
plt.figure(figsize = (10, 6))
plt.title('XOR in new dimensions')
for i in range(len(df)):
    plt.scatter(output_matrix[i][0],output_matrix[i][1],color = ListedColormap(('red', 'green'))(y[i]))
plt.show()
```
